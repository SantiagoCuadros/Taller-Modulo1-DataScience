### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Promedio
total = 0
for valor in esperanzas_de_vida:
    total += valor
promedio = total / len(esperanzas_de_vida)
print(f"Promedio: {promedio:.2f}")

# Máximo
maximo = esperanzas_de_vida[0]
for valor in esperanzas_de_vida:
    if valor > maximo:
        maximo = valor
print(f"Máximo: {maximo}")

# Mínimo
minimo = esperanzas_de_vida[0]
for valor in esperanzas_de_vida:
    if valor < minimo:
        minimo = valor
print(f"Mínimo: {minimo}")
````

### 1.3 — Funciones

Las funciones nos permiten reutilizar lógica.

**Ejercicio 1.3** — Crea una función llamada `clasificar_esperanza_vida` que reciba un número (años) y retorne:
- `"Alta"` si es >= 75
- `"Media"` si es entre 60 y 74.9
- `"Baja"` si es < 60

Prueba la función con al menos 5 valores distintos e imprime los resultados.

````python
def clasificar_esperanza_vida(años):
    if años >= 75:
        return "Alta"
    elif años >= 60:
        return "Media"
    else:
        return "Baja"

# Pruebas con 5 valores
valores_prueba = [80.0, 72.3, 55.4, 76.5, 49.3]
for v in valores_prueba:
    print(f"{v} años = {clasificar_esperanza_vida(v)}")
````

### 1.4 — Listas + Funciones + Bucles (integrador)

**Ejercicio 1.4** — Usa la función creada en 1.3 y la lista del ejercicio 1.1 para:
1. Crear una **nueva lista** llamada `clasificaciones` que contenga la clasificación de cada esperanza de vida
2. Imprimir cada par: `(valor, clasificación)` usando un bucle `for`

````python
# Nueva lista con clasificaciones
clasificaciones = []
for valor in esperanzas_de_vida:
    clasificaciones.append(clasificar_esperanza_vida(valor))

# Imprimir cada par
for valor, clasificacion in zip(esperanzas_de_vida, clasificaciones):
    print(f"{valor} años = {clasificacion}")
````

### 2.1 — Carga y exploración inicial

````python
import pandas as pd

# Cargar el dataset
df = pd.read_csv("salud_mundial.csv")

# 1. Dimensiones
print("Filas y columnas:", df.shape)

# 2. Tipos de datos
print("\nTipos de datos:")
print(df.dtypes)

# 3. Valores nulos
print("\nValores nulos por columna:")
print(df.isnull().sum())

# 4. Primeras 8 filas
print("\nPrimeras 8 filas:")
df.head(8)
````

### 2.2 — Estadísticas descriptivas

🔧 **Ejercicio 2.2** — Usa `describe()` para obtener estadísticas generales.  
Luego responde:
1. ¿Cuál es la esperanza de vida promedio mundial?
2. ¿Cuál es el gasto máximo y mínimo en salud?
3. ¿Cuál es la mediana de mortalidad infantil?

````python
# Estadísticas generales
print(df.describe())

# 1. Esperanza de vida promedio
print(f"\nEsperanza de vida promedio: {df['esperanza_vida'].mean():.2f} años")

# 2. Gasto máximo y mínimo
print(f"Gasto máximo en salud: {df['gasto_salud_usd'].max():.2f} USD")
print(f"Gasto mínimo en salud: {df['gasto_salud_usd'].min():.2f} USD")

# 3. Mediana de mortalidad infantil
print(f"Mediana mortalidad infantil: {df['mortalidad_infantil'].median():.2f}")
````

### 2.3 — Limpieza de datos

🔧 **Ejercicio 2.3** — Limpia los valores nulos:  
1. Rellena los nulos de columnas numéricas con la **mediana** de cada columna
2. Verifica que ya no haya nulos
3. ¿Cuántos registros tenía el dataset original vs después de limpiar?

````python
# Registros originales
print(f"Registros originales: {len(df)}")
print(f"Nulos antes de limpiar:\n{df.isnull().sum()}")

# Rellenar nulos numéricos con la mediana
columnas_numericas = df.select_dtypes(include='number').columns
for col in columnas_numericas:
    df[col].fillna(df[col].median(), inplace=True)

# Verificar
print(f"\nNulos después de limpiar:\n{df.isnull().sum()}")
print(f"Registros después de limpiar: {len(df)}")
````

### 2.4 — Filtros

🔧 **Ejercicio 2.4**:
1. Filtra los países de la región **"Americas"**
2. Filtra los países con esperanza de vida **mayor a 80 años** — ¿cuántos son? ¿De qué regiones son?

4. ¿Cuántos países tienen vacunación por encima del 90%?

### 2.5 — Agrupaciones

🔧 **Ejercicio 2.5** — Usa `groupby()` para:
1. Calcular la **esperanza de vida promedio** por `region`
2. Calcular el **gasto en salud promedio** por `nivel_ingresos`
3. Calcular el **número de países** por región
4. ¿Qué región tiene la mayor esperanza de vida promedio? ¿Y la menor?

````python
# 1. Esperanza de vida promedio por región
ev_por_region = df.groupby('region')['esperanza_vida'].mean().round(2)
print("Esperanza de vida promedio por región:")
print(ev_por_region)

# 2. Gasto en salud promedio por nivel de ingresos
gasto_por_nivel = df.groupby('nivel_ingresos')['gasto_salud_usd'].mean().round(2)
print("\nGasto en salud promedio por nivel de ingresos:")
print(gasto_por_nivel)

# 3. Número de países por región
paises_por_region = df.groupby('region')['pais'].count()
print("\nNúmero de países por región:")
print(paises_por_region)

# 4. Mayor y menor esperanza de vida
print(f"\nRegión con MAYOR esperanza de vida: {ev_por_region.idxmax()} ({ev_por_region.max()} años)")
print(f"Región con MENOR esperanza de vida: {ev_por_region.idxmin()} ({ev_por_region.min()} años)")
````

### 3.3 — Buenas prácticas

**¿Qué archivos debería incluir un `.gitignore` en un proyecto de Data Science? Menciona al menos 3 y explica por qué.**

> 1. pycache :carpeta generada automáticamente por Python que no aporta al proyecto y ocupa espacio.
> 2. .ipynb_checkpoints  :carpeta de checkpoints automáticos de Jupyter, no es necesario subirla al repositorio.
> 3. *.env :archivos con variables de entorno que pueden contener contraseñas o tokens privados que no deben exponerse públicamente.

**¿Cuál es la diferencia entre `git add .` y `git add nombre_archivo`?**

> git add . :agrega TODOS los archivos modificados o nuevos del directorio actual al staging area de una sola vez. En cambio, git add nombre_archivo: agrega únicamente el archivo específico que indiques. El primero es más rápido pero menos preciso; el segundo te da más control sobre qué cambios incluir en cada commit.
